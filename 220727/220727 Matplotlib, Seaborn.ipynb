### matplotlib의 구조
- 매트랩에서 사용하는 다양한 기능을 사용하게 해주는 도구
- 엑셀에서 사용하는 정형화된 차트나 그래프를 작성할 수 있게 해준다.

```python
import matplotlib.pyplot as plt

X = range(100)
Y = range(100)
plt.plot(X,Y)
```

```python
import numpy as np

X_1 = range(100)
Y_1 = [np.cos(value) for value in X_1] #코사인

X_2 = range(100)
Y_2 = [np.sin(value) for value in X_2] #사인

plt.plot(X_1, Y_1)
plt.plot(X_2, Y_2)
plt.show()
```

```python
# subplots 함수를 통해 하나의 figure에 여러 그래프를 그릴수 있다.

fig, ax = plt.subplots() #figure와 axes 객체 할당

X_1 = range(100)
Y_1 = [np.cos(value) for value in X_1] #코사인

ax.plot(X_1,Y_1) 
ax.set(title = 'cos graph', 
      xlabel = 'X',
      ylabel = 'Y');

plt.show()
```

```python
fig = plt.figure()
fig.set_size_inches(10,10) #fihure의 크기 지정

ax_1 = fig.add_subplot(1,2,1) #그래프ㅡ 위치 지정
ax_2 = fig.add_subplot(1,2,2)

ax_1.plot(X_1,Y_1,c="b") #그래프 축과 색깔 살정
ax_2.plot(X_2,Y_2,c="g")
plt.show()
```

#### 서브플롯 행렬

- 축을 여러개 만들때 일반적으로 figure 객체에서 add_subplot 함수를 사용하거나 plot 객체에서 subplots 함수를 사용하여 객체공간을 확보해야함

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
print(ax) #넘파이 배열임
```

```python
x = np.linspace(-1,1,100)
y_1 = np.sin(x)
y_2 = np.cos(x)
y_3 = np.tan(x)
y_4 = np.exp(x)

fig, ax = plt.subplots(2,2)


ax[0,0].plot(x,y_1)
ax[1,0].plot(x,y_2)
ax[0,1].plot(x,y_3)
ax[1,1].plot(x,y_4)
```

```python
ax1 = plt.subplot(321)
plt.plot(x,y_1)
ax1 = plt.subplot(322)
plt.plot(x,y_1)
ax1 = plt.subplot(312)
plt.plot(x,y_1)
ax1 = plt.subplot(325)
plt.plot(x,y_1)
ax1 = plt.subplot(326)
plt.plot(x,y_1)

plt.show()
```

```python
X_1 = range(100)
Y_1 = [value for value in X_1]

X_2 = range(100)
Y_2 = [value + 100 for value in X_2]

plt.plot(X_1, Y_1,  color= "#75C758")
plt.plot(X_2, Y_2, c="c")

plt.show()
```

```python
plt.plot(X_1, Y_1,  color= "#75C758",
        linestyle ="dashed") #점선
plt.plot(X_2, Y_2, c="c",
        ls="dotted") #도트선

plt.show()
```

```python
plt.plot(X_1, Y_1,  color= "#75C758",
        linestyle ="dashed")
plt.plot(X_2, Y_2, c="c",
        ls="dotted")

plt.title("Two lines")
plt.show()
```

```python
fig = plt.figure()
fig.set_size_inches(10,2)

ax_1 = fig.add_subplot(1,2,1)
ax_2 = fig.add_subplot(1,2,2)

ax_1.plot(X_1, Y_1, c ="b")
ax_1.set_title("Figure 1")

ax_2.plot(X_2, Y_2, c ="g")
ax_2.set_title("Figure 2")

plt.show()
```

```python
plt.plot(X_1, Y_1,  color= "#75C758",
        linestyle ="dashed",
        label = '라인_1') #범례 설정
plt.plot(X_2, Y_2, c="c",
        ls="dotted",
        label = '라인_2')
plt.legend(              #범례 설정하는 함수
    shadow=True,
    loc = "upper right") #범례 위치설정
plt.title("$y_1 = ax_1+b$") #$$ 사이에 알파벳들어가면 이텔릭체가 됨
plt.xlabel('$x_1line$') #_하면 한글자만 작은글자로 들어감
plt.ylabel('$y_{line}$') # 중괄호로 _뒤를 묶으면 묶은만큼 작아짐
plt.show()
```

```python
# 산점도 그래프

data_1 = np.random.rand(512,2)
data_2 = np.random.rand(512,2)

plt.scatter(data_1[:,0],
           data_1[:,1],
           c="b", marker="x") #maker : 점모양 설정

plt.scatter(data_2[:,0],
           data_2[:,1],
           c="r", marker="o")
```

```python
N = 50 #숫자를 전부 맞춰야함
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)

area = np.pi*(
15* np.random.rand(N)) **2
plt.scatter(x,y,
           s=area, c = colors, #s는 size
           alpha=0.5) #투명도

plt.show()
```

```python
#막대그래프

data = [[5.,25.,50.,20.],
       [4.,23.,51.,17.],
       [6.,22.,52.,19.]]

X = np.arange(0,8,2) #좌표 시작점 x축의 간격 지정

plt.bar(X + 0.00, data[0], color = 'b', width = 0.5)#띄워주기위해 간격다르게 지정 억지로 만든거
plt.bar(X + 0.50, data[1], color = 'g', width = 0.5)
plt.bar(X + 1.00, data[2], color = 'r', width = 0.5)

plt.xticks(X+0.50,('가','나','다','라'))

plt.show()
```

```python
data = np.array([[5.,25.,50.,20.],
       [4.,23.,51.,17.],
       [6.,22.,52.,19.]])

color_list = ['b','g','r']
data_label = ["A","B","C"]
X = np.arange(data.shape[1])

for i in range(3):
    plt.bar(X, data[i], #data[i]
    bottom = np.sum(    #bottom은 y축 그래프의 
    data[:i], axis = 0), #누적합 그래프
    color = color_list[i],
    label = data_label[i])
    
plt.legend()
plt.show()
```

#### Seaborn

- matplotlib보다 좀더 이쁨


```python
import seaborn as sns

fmri = sns.load_dataset("fmri")
sns.set_style("whitegrid") #스타일 적용 다양한 스타일이 있다.
sns.lineplot(x="timepoint", y="signal", data = fmri)
```

```python
sns.lineplot(x='timepoint',y='signal',hue='event',data=fmri) #이벤트 범주에 속한 데이터만 시각화
```

#회귀 그래프
- 단순한 선그래프에도 시본에서는 다양한 그래프 사용가능


```python
#회귀선 추가가능
tips = sns.load_dataset("tips")
sns.regplot(x='total_bill', y='tip', data=tips, x_ci = 95)
#x_ci -> 신뢰구간 비율을 나타내는 매개변수
```

```python
# 산점도
tips = sns.load_dataset("tips")
sns.scatterplot(x='total_bill', y='tip', hue='time', data=tips)
```

```python
#빈도그래프
tips = sns.load_dataset("tips")
sns.countplot(x='smoker', hue='time', data=tips)
```

```python
#막대그래프
sns.barplot(x="day", y='total_bill', data= tips)
```

```python
#바이올린 플롯과 스웜플롯

sns.violinplot(x='day',y='total_bill',hue='smoker',
              data=tips, palette="muted")
```

```python
sns.swarmplot(x='day',y='total_bill',hue='smoker',
              data=tips, palette="muted") #palette <- 색깔
```

```python
#기본적으로 col과 row에 열을 넣으면 데이터의 종류만큼 m x n이 됨
g = sns.FacetGrid(tips, col='time', #시간은 런치 디너
                 row='sex') #성별은 남녀 두개라서 (2,2)가 된다
```

```python
g = sns.FacetGrid(tips,
                 col='time',
                 row='sex')

g.map(sns.scatterplot,
     "total_bill","tip")
```

```python
g = sns.FacetGrid(tips, col='time',
                 row = "sex")

g.map_dataframe(sns.histplot,
               x="total_bill")
```
