```python
#멀티인덱스일 떄 파일불러올때는 칼럼을 선택해서 데이터프레임을 불러옴 * 옵션 usecols - '원하는 칼럼'
```

```python
import pandas as pd

df_car = pd.read_excel("c:/py_data/car_in_seoul.xls",
                      header=1, #엑셀 2번째 행을 헤더로 가져옴
                      usecols = 'A, B,C,G,K,O,S') #원하는 열만 선택해서 불러옴

df_car.head()
```

```python
df_par = pd.read_excel("c:/py_data/par_in_seoul.xls",
                      header=2, #엑셀 3번째 행을 헤더로 가져옴
                      usecols = 'A,B,C,I,O,W') #원하는 열만 선택해서 불러옴
df_par.head()
```

```python
# 필드명 변경
df_car.rename(columns={ df_car.columns[0] : '년도', #혹시 모를 문자가 섞여 있을수도 있어서, 주소로 설정
                      df_car.columns[1] : '자치구',
                      df_car.columns[2] : '합계',
                      df_car.columns[3] : '승용차',
                      df_car.columns[4] : '승합차',
                      df_car.columns[5] : '화물차',
                      df_car.columns[6] : '특수차'},
              inplace=True)
df_car.head()
```

```python
# 필드명 변경
df_par.rename(columns={ df_par.columns[0] : '년도',
                      df_par.columns[1] : '자치구',
                      df_par.columns[2] : '합계',
                      df_par.columns[3] : '노상',
                      df_par.columns[4] : '노외',
                      df_par.columns[5] : '건축물부설'},
              inplace=True)
df_par.head()
```

```python
# 승용차 데이터의 불필요한 행을 삭제합니다.
#(서울시 기타) 행 삭제


df_car.drop([0],inplace=True) # 자동차 서울시 제거
df_car.drop([26],inplace=True) # 자동차 기타 제거
df_par.drop([0],inplace=True) # 주차장 서울시 제거
```

```python
df_car
```

```python
df_car.info() # data의 전체적인 정보를 제공
```

```python
# 컬럼 데이터 형식 변환 astype(바꿀칼럼)
df_car['합계'] = df_car['합계'].astype(int) 
```

```python
df_car.info()
```

```python
# df_card에 승용차, 승합차 화물차 비율 추가
# 새로운칼럼생성
df_car['승용차비율'] = df_car['승용차']/df_car['합계']*100
df_car['승합차비율'] = df_car['승합차']/df_car['합계']*100
df_car['화물차비율'] = df_car['화물차']/df_car['합계']*100
df_car.head(10)
```

```python
df_par
```

```python
# 노상비율, 노외비율, 건축물 비율 계산필드 추가

df_par['노상비율'] = df_par['노상']/df_par['합계']*100
df_par['노외비율'] = df_par['노외']/df_par['합계']*100
df_par['건축물비율'] = df_par['건축물부설']/df_par['합계']*100
df_par.head(10)
```

```python
# df_car1 = df_car 이렇게 카피본을 만들면 하나의 변수에서 값을 수정하면 다른 변수에서도 값이 수정됨 같은곳을 가르킴 *주의

# 데이터프레임복사 함수
df_car1 = df_car.copy()
df_par1 = df_par.copy()
```

```python
#df_cal1에 불필요한 필드
# 승용차 승합차 화물차 특수차 삭제
del df_car1['승용차']
del df_car1['승합차']
del df_car1['화물차']
del df_car1['특수차']
df_car1.head()
```

```python
del df_par1['노상']
del df_par1['노외']
del df_par1['건축물부설']
df_par1.head()
```

```python
# 차량등록자료와 주차장 등록 자료를
# merge으로 합치기를 실행합니다. (조인)
# 데이터 프레임 명은 data seoul`
# 자치구 기준
data_seoul = pd.merge(df_car1, df_par1,
                     on = '자치구')
data_seoul.head()
```

```python
# 중복 칼럼 이름 변경(합계는 필요함) and 삭제(년도는 1개만)
data_seoul.rename(columns={data_seoul.columns[0]:"년도",
                          data_seoul.columns[2]:"차량합계",
                          data_seoul.columns[7]:"주차장합계"},
                 inplace = True)
data_seoul.head()
```

```python
del data_seoul['년도_y']
data_seoul.head()
```

```python
# data_seoul 에서 자치구를 인덱스 키로 변경합니다
```

```python
data_seoul.set_index("자치구", inplace=True)
data_seoul.head()
```
