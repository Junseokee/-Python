#### 판다스
##### 판다스의 개념
- 판다스는 데이터 분석 라이브러리로 엑셀과 유사함
- 데이터 테이블을 다루기 위한 도구로 적합
- 판다스를 다루기위해 넘파이 사용
- 데이터 테이블형태인 2차원 데이터 사용
- 판다스에서 테이블 객체를 데이터 프레임이라 하고 각각의 열 데이터를 시리즈라 한다.

##### 시리즈 객체
  - 시리즈 객체는 피쳐 벡터와 같은 개념으로 사용, 일반적으로 하나의 피쳐 데이터를 포함하는 형태
  - 시리즈 객체는 데이터프레임안에 포함되기도 하지만 list, dict, ndarray 등과 같은 타입도 시리즈 객체가 된다.

```python
from pandas import Series, DataFrame
list_data = [1,2,3,4,5]
list_name = ['a','b','c','d','e']
example_obj = Series(data=list_data, index=list_name)
example_obj
```

#### 그룹별 집계의 개념

- 그룹별 집계가 가능하다 그루핑
- 그룹바이는 데이터로부터 동일한 객체를 가진 데이터만 뽀ㅃ아 기술 통계 데이터를 추출하는 기능

        - 분할(split), 적용(apply), 결합(combine)

```python
ipl_data = {'Team' : ['Riders','Riders','Devils','Devils','Kings','Kings','Kings','Kings',
                     'Riders','Royals','Royals','Riders'],
           'Rank' : [1,2,2,3,3,4,1,1,2,4,1,2],
           'Year' : [2014, 2015, 2014, 2015, 2014, 2015, 2016, 2017,
                    2016, 2014, 2015, 2017],
           'Points' : [876, 789, 863, 673, 741, 812, 
                      756, 788, 694, 701, 804, 690]}

df = pd.DataFrame(ipl_data)
df
```

```python
df.groupby("Team")["Points"].sum() #팀별 점수의 합계를 구함
```

```python
multi_groupby = df.groupby(["Team",'Year'])['Points'].sum() #인덱스가 두개 이상 같은팀당 연도별 포인트를 합산
multi_groupby 
```

```python
multi_groupby = df.groupby(["Team",'Year'])["Points"].sum()
multi_groupby.index
```

```python
multi_groupby["Devils":'Kings']
```

- unstack 함수를 사용하여 기존 인덱스를 기준으로 묶인 데이터에서 두 번째 인덳를 열로 변화시켜 엑셀의 피봇테이블과 비슷한 형태로 데이터를 볼 수 있다.

```python
multi_groupby.unstack()
```

- swaplevel 함수를 이용하면 인덱스간의 레벨을 변경 할 수 있다.

```python
multi_groupby.swaplevel().sort_index() # 연도와 팀 인덱스 변경 그 후 sort_index 함수를 통해 정렬
```

```python
multi_groupby.sum(level=0) # level은 0번째를 의미 팀별, 1이되면 연도
```

```python
multi_groupby.sum(level=1)
```

### 그룹화된 상태

- 분할 -> 적용 -> 결합으로 이루어지는 grouby 함수 사용에서 분할만 적용한 상태로 데이터를 다룰 수 있는데, 이를 grouped상태라 부른다.
- 그룹화된 상태에서 get_group 함수를 사용하면 해당 키값을 시준으로 분할된 데이터프레임 객체를 확인 할 수 있다.

```python
grouped = df.groupby("Team")
grouped.get_group("Riders")
```

그룹화된 상태의 데이터프레임에 
- 집계, 변환, 필터 등 사용 가능

```python
grouped.agg(min) # 팀을 기준으로 모든 컬럼에서 가장 작은 값이 나옴 / agg : 다양한 함수를 그대로 적용 시킬수 있다.
```

```python
grouped.agg(np.mean)
```

```python
grouped.transform(max) #팀을 기준으로 가장 큰값으로 값을 변경함
```

```python
score = lambda x: (x-x.mean()) / x.std() #정규화 공식
grouped.transform(score)
```

```python
df.groupby('Team').filter(lambda x: len(x) >= 3) # x(Team)길이가 3개 이상인것만 필터링 
```

```python
df.groupby('Team').filter(
lambda x:x["Points"].max()>810) #최대점수가 810이 넘는 팀의 기록만 출력
```

### 병합 join
- 두개의 데이터를 특정한 기준을 갖고 하나로 통합하는 작업

```python
raw_data = {
    'subject_id' : ['1','2','3','4','5','7','8','9','10','11'],
    'test_score' : [51,15,15,61,16,14,15,1,61,16]}

df_left = pd.DataFrame(raw_data, columns = ['subject_id', 'test_score'])
df_left
```

```python
raw_data = {
    'subject_id' : ['4','5','6','7','8'],
    'first_name' : ['Billy', 'Brian', 'Bran','Bryce', 'Betty'],
    'last_name' : ['Bonder','Black', 'Balwner', 'Brice','Btisan']}

df_right = pd.DataFrame(raw_data, columns = ['subject_id', 'first_name', 'last_name'])
df_right
```

```python
pd.merge(left=df_left, right=df_right, how="inner", on='subject_id') #왼쪽 오른쪽 데이터프레임 선택, how 어떤 조인인지, on 기준되는 칼럼
```

```python
pd.merge(df_left, df_right, how="right", on='subject_id')
```

```python
pd.merge(df_left, df_right, how="left", on='subject_id')
```

```python
pd.merge(df_left, df_right, how="outer", on='subject_id')
```

```python
pd.merge(df_left, df_right, on='subject_id') # how의 디폴트값이 inner join
```

```python
df_left.index = df_left.subject_id
del df_left['subject_id']
df_right.index = df_left.subject_id
del df_right['subject_id']

pd,merge(df.left,df.right,on='subject_id')
```
