```python
#함수화 진행

def getTitleClien(page): #input page = 출력할 페이지수 
    titleList = [] #반복문 밖에서 공리스트 생성 안에있으면 계속 초기화 됨
    for pageNum in range(page): 
        urlStr = 'https://www.clien.net/service/board/sold?&od=T31&category=0&po='+str(pageNum) 

        resp = requests.get(urlStr)
        htmlText = resp.text

        titleFirst = 'data-role="list-title-text" title="'
        titleLast = '">' 
        titleLastindex = 0

        while True: 
            try:
                titleFirstindex = htmlText.index(titleFirst,titleLastindex) + len(titleFirst)
            except ValueError: 
                break
            titleLastindex = htmlText.index(titleLast,titleFirstindex)
            titleStr = htmlText[titleFirstindex:titleLastindex] 
            titleList.append(titleStr) #리스트에 제목을 추가
    return titleList #최종 output
```
